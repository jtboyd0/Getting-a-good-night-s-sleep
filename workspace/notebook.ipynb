Your client, SleepInc, has shared anonymized sleep data from their hot new sleep tracking app SleepScope. As their data science consultant, your mission is to analyze the lifestyle survey data with Python to discover relationships between exercise, gender, occupation, and sleep quality. ## The data: sleep_health_data.csv

SleepInc has provided you with an anonymized dataset of sleep and lifestyle metrics for 374 individuals. This dataset contains average values for each person calculated over the past six months. The data is saved as `sleep_health_data.csv`.

The dataset includes 13 columns covering sleep duration, quality, disorders, exercise, stress, diet, demographics, and other factors related to sleep health. 

| Column | Description |
|---------|----------------------------------------|  
| `Person ID` | An identifier for each individual. |
| `Gender` | The gender of the person (Male/Female). |  
| `Age` | The age of the person in years. |
| `Occupation` | The occupation or profession of the person. |
| `Sleep Duration (hours)` | The average number of hours the person sleeps per day. |
| `Quality of Sleep (scale: 1-10)` | A subjective rating of the quality of sleep, ranging from 1 to 10. |
| `Physical Activity Level (minutes/day)` | The average number of minutes the person engages in physical activity daily. |  
| `Stress Level (scale: 1-10)` | A subjective rating of the stress level experienced by the person, ranging from 1 to 10. |
| `BMI Category` | The BMI category of the person (e.g., Underweight, Normal, Overweight). |
| `Blood Pressure (systolic/diastolic)` | The average blood pressure measurement of the person, indicated as systolic pressure over diastolic pressure. |
| `Heart Rate (bpm)` | The average resting heart rate of the person in beats per minute. |
| `Daily Steps` | The average number of steps the person takes per day. |
| `Sleep Disorder` | The presence or absence of a sleep disorder in the person (None, Insomnia, Sleep Apnea). |

```python
# Imports
import pandas as pd

sleep_health_data = pd.read_csv('sleep_health_data.csv')

display(sleep_health_data.info(), sleep_health_data.head(), sleep_health_data.isna().sum())
```

```python
# Explore the discrete and continuous columns in the dataframe

sleep_health_data[sleep_health_data.columns[1:]].describe()
```

```python
# Explore the categorical columns in the dataframe
# Gender, Occupation, Sleep Disorder

categorical_columns = ['Gender', 'Occupation', 'Sleep Disorder', 'BMI Category']
for column in categorical_columns:
    display(sleep_health_data[column].value_counts())
    
sleep_health_data[categorical_columns].value_counts()
```

## Which occupation has the lowest average sleep duration?

```python
# Group rows by occupation and find the minimum average sleep duration
average_sleep_duration_occ = sleep_health_data.groupby('Occupation')['Sleep Duration'].mean()
lowest_sleep_avg = average_sleep_duration_occ.min()
lowest_sleep_occ = average_sleep_duration_occ.idxmin()

display(f'The occupation with the lowest average sleep duration is {lowest_sleep_occ} with {lowest_sleep_avg} hours of sleep.')
```

## Which occupation has the lowest average sleep quality?

```python
# Group rows by occupation and find the minimum average sleep quality
average_sleep_quality_occ = sleep_health_data.groupby('Occupation')['Quality of Sleep'].mean()
lowest_sleep_quality_avg = average_sleep_quality_occ.min()
lowest_sleep_quality_occ = average_sleep_quality_occ.idxmin()

display(f'The occupation with the lowest average sleep quality (A subjective rating of the quality of sleep, ranging from 1 to 10) is {lowest_sleep_quality_occ} with a rating of {lowest_sleep_quality_avg}.')
```

## Did the occupation with the lowest sleep duration also have the lowest sleep quality?

```python
same_occ = (lowest_sleep_occ == lowest_sleep_quality_occ)

if same_occ:
    display(f'It is true, the occupation {lowest_sleep_occ} was found to have the lowest sleep duration on average and also reported, on average, the lowest quality of sleep.')
else:
    display(f'It is false, the occupation with the lowest sleep duration on average ({lowest_sleep_occ}) did not, on average, report to have the lowest quality of sleep. The lowest quality of sleep on average was found in people that help the occupation of {lowest_sleep_quality_occ}')
```

## What is the ratio of app users in each BMI category that have been diagnosed with insomnia?

```python
# Group data frame by BMI Category and Sleep disorder
grouped_bmi_disorder = sleep_health_data.groupby(['BMI Category', 'Sleep Disorder']).size()

# Create a list of unique values in BMI Category column
unique_bmi_values = sleep_health_data['BMI Category'].unique()

# Use comprehension to create a dictionary that contains the Bmi category and the ratio of insomniacs found in that category
bmi_insomnia_ratios = {cat : round(grouped_bmi_disorder[cat]['Insomnia']/grouped_bmi_disorder[cat].sum(), 2) for cat in unique_bmi_values}

bmi_insomnia_ratios
```
